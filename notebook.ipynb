```python
import gym 
from stable_baselines import ACER
from stable_baselines.common.vec_env import DummyVecEnv 
from stable_baselines.common.evaluation import evaluate_policy 
```

```python
environment_name = 'LunarLander-v2'
```

#### 1. Test Random Environment

```python
env = gym.make(environment_name)

episodes = 3
for episode in range(1, episodes+1):
    state = env.reset()
    done = False 
    score = 0 

    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score += reward 
    print(f'Episode:{episode}, Score:{score}')
env.close()
```

#### 2. Build and Train the Model

```python
import gym

gym.__version__
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env]) # wrap the environment into the DummyVecEnv
model = ACER('MlpPolicy', env, verbose=1) #MLP -> multi-layer-preceptron (nn in the background)
```

```python
model.learn(total_timesteps=100_000)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
env.close()
```

```python
model.save('ACER_MODEL')
```

```python
model = ACER.load('ACER_MODEL', env=env)
```

```python
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    env.render()

env.close()
```
